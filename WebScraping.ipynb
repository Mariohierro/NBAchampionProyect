## 1. Obtenemos los datos ofensivos

```python
import requests
#import os
#import shutil
import pandas as pd
```

Creamos un rango para todos los años de los que extraeremos datos

```python
years = list(range(2013,2024))
```

Creamos una carpeta "Play_offs" en donde guardaremos toda la informacion de la url en la que se encuentra la tabla que queremos obtener. Así no tendremos que acceder a la url continuamente, ya tendremos esa informacion descargada aquí.

Creamos el siguiente bucle para obtener y guardar la informacion año por año

```python
url_start = "https://www.basketball-reference.com/playoffs/NBA_{}.html"

for year in years:
    url = url_start.format(year)
    
    data = requests.get(url)
    
    with open("PlayOffs/{}.html".format(year), "w+", encoding="utf-8") as f: #cargamos la informacion de cada año
        f.write(data.text)                                              # en la carpeta creada
```

#### Creamos ahora el siguiente bucle para obtener la tabla de cada año sin tener que hacerlo uno por uno

```python
dfs = [] #creamos una lista vacía donde juntaremos todas las tablas
for year in years:
    if year == 2023:      #añadimos esta condicion ya que la tabla del 2023 es algo diferente y tenemos que
        with open("PlayOffs/{}.html".format(year), encoding="utf8") as f:     # modificar algunas columnas antes 
            page = f.read()                                              # de juntarla con todas las demas
    
        soup = BeautifulSoup(page, 'html.parser')  # accedemos a la url (en este caso la carpeta con la informacion)
        tabla_PO = soup.find_all(id="per_game-team")  # nos quedamos con la informacion de la tabla

        tabla_year = pd.read_html(str(tabla_PO))[0]   # nos quedamos con la parte que nos interesa de la tabla y la convertimos a fotmato lista 
        tabla_year.drop(columns='Rk', inplace=True)
        tabla_year['Team'] = tabla_year['Team'] + ' ' + str(year)       # borramos la columna 'Rk'
        tabla_year = tabla_year.rename(columns={'Team': 'Team'})   #Esto lo habiamos hecho ya que para el año 2023 la columna 'Team'
    else:                                                       # se llamaba en un principio 'Tm'
        with open("PlayOffs/{}.html".format(year), encoding="utf8") as f: 
            page = f.read()                                

        soup = BeautifulSoup(page, 'html.parser')         # Bucle para el resto de años
        tabla_PO = soup.find_all(id="per_game-team") 

        tabla_year = pd.read_html(str(tabla_PO))[0]
        tabla_year.drop(columns='Rk', inplace=True)
        tabla_year['Team'] = tabla_year['Team'] + ' ' + str(year)       
        tabla_year = tabla_year.rename(columns={'Team': 'Team'})
    dfs.append(tabla_year)  # juntamos todas las tablas ¿ sin formato (adecuado) ?
```

## 3. Obtenemos los datos de la posicion de cada equipo en temporada regular

Ya tenemos los datos de los equipos desde 2013 hasta 2015 de ambas conferencias, por lo que solo nos faltan desde 2016

### 3.1. Datos de equipos del ESTE

```python
url_start = "https://www.basketball-reference.com/leagues/NBA_{}_standings.html"

for year in years:
    url = url_start.format(year)
    
    data = requests.get(url)
    
    with open("posicion/{}.html".format(year), "w+", encoding="utf-8") as f: #cargamos la informacion de cada año
        f.write(data.text)                                              # en la carpeta creada
```

```python

dfs = [] #creamos una lista vacía donde juntaremos todas las tablas
dfs2 = []
for year in years:   
    with open("posicion/{}.html".format(year), encoding="utf8") as f:     
        page = f.read()                                              
    
    soup = BeautifulSoup(page, 'html.parser')  # accedemos a la url (en este caso la carpeta con la informacion)
    posiciones_este = soup.find_all(id="confs_standings_E") # nos quedamos con la informacion de la tabla
    posiciones_oeste = soup.find_all(id="confs_standings_W")
    
    equipo_posicion_este = []
    equipo_posicion_oeste = []

# OBTENEMOS LOS DATOS DE LA TABLA DEL ESTE
    for tabla in posiciones_este:
        for fila in tabla.find_all('tr')[1:]:
    # Encuentra todas las celdas de la fila
            celdas = fila.find_all('th')
                
    # Verifica si hay suficientes celdas en la fila
            if len(celdas) >= 1:
        # Extrae los datos de la columnas 1
                equipo_posicion_este.append(celdas[0].text.strip())
     
    posiciones_este = pd.DataFrame({'Equipo_pos' : equipo_posicion_este})
    posiciones_este['año'] = str(year)
   
    dfs.append(posiciones_este)  # juntamos todas las tablas ¿ sin formato (adecuado) ?

    
# OBTENEMOS LOS DATOS DE LA TABLA DEL OESTE
    for tabla in posiciones_oeste:
        for fila in tabla.find_all('tr')[1:]:
    # Encuentra todas las celdas de la fila
            celdas = fila.find_all('th')
    
    # Verifica si hay suficientes celdas en la fila
            if len(celdas) >= 1:
        # Extrae los datos de la columnas 1
                equipo_posicion_oeste.append(celdas[0].text.strip())
     
    posiciones_oeste = pd.DataFrame({'Equipo_pos' : equipo_posicion_oeste})
    posiciones_oeste['año'] = str(year)
    
    dfs2.append(posiciones_oeste)  # juntamos todas las tablas ¿ sin formato (adecuado) ?
```

```python
df_posiciones_este_16_22 = pd.concat(dfs)
df_posiciones_este_16_22
```

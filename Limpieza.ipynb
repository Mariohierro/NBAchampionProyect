```python
filas_a_borrar = list(range(160,176))
df_NBA = df_NBA.drop(filas_a_borrar)
df_NBA = df_NBA.drop(columns= 'G')
del df_NBA['MP']
```

```python
df_NBA.tail()
```

## 2. Limpiamos los datos de la posicion

### 2.1. Datos del ESTE

```python
df_posiciones_este_16_22 = pd.read_csv('df_posiciones_este_16_22.csv', sep=',')
df_posiciones_oeste_16_22 = pd.read_csv('df_posiciones_oeste_16_22.csv', sep=',')
```

```python
df_posiciones_este_16_22
```

```python
df_posiciones_este_16_22['Equipo_pos'] = df_posiciones_este_16_22['Equipo_pos'].str.split('*').str[0]

df_posiciones_este_16_22.tail(19)
```

```python
df_posiciones_este_16_22.shape
```

Borramos ahora los equipos que quedaron entre el 9 y el 15

```python
df_posiciones_este_16_22 = df_posiciones_este_16_22[~df_posiciones_este_16_22['Unnamed: 0'].between(8, 14)]

df_posiciones_este_16_22
```

```python
df_posiciones_este_16_22['posicion'] = df_posiciones_este_16_22['Unnamed: 0'] + 1  # Añadimos la columna posicion adecuadamente
```

```python
del df_posiciones_este_16_22['Unnamed: 0'] # borramos columnas innecesarias
df_posiciones_este_16_22.shape
```

```python
df_posiciones_este_16_22['Equipo_pos'] = df_posiciones_este_16_22['Equipo_pos'] + ' ' + df_posiciones_este_16_22['año'].astype(str)  #añadimos el año a la columna equipo
df_posiciones_este_16_22 = df_posiciones_este_16_22.rename(columns={'Equipo_pos':'Team'})   #renombramos la columna equipo para poder juntarla posteriormente con la tabla general
del df_posiciones_este_16_22['año']                                                        # borramos la columna año
df_posiciones_este_16_22
```

```python
#Corregimos algunos datos que se han cargado incorrectamente
df_posiciones_este_16_22.loc[df_posiciones_este_16_22['Team'] == 'Miami Heat 2023', 'posicion'] = 8
df_posiciones_este_16_22.loc[df_posiciones_este_16_22['Team'] == 'Atlanta Hawks 2023', 'posicion'] = 7
df_posiciones_este_16_22.tail()
```

### Datos del OESTE

```python
df_posiciones_oeste_16_22['Equipo_pos'] = df_posiciones_oeste_16_22['Equipo_pos'].str.split('*').str[0]
df_posiciones_oeste_16_22 = df_posiciones_oeste_16_22[~df_posiciones_oeste_16_22['Unnamed: 0'].between(8, 14)]
df_posiciones_oeste_16_22['posicion'] = df_posiciones_oeste_16_22['Unnamed: 0'] + 1  # Añadimos la columna posicion adecuadamente
del df_posiciones_oeste_16_22['Unnamed: 0'] # borramos columnas innecesarias
df_posiciones_oeste_16_22['Equipo_pos'] = df_posiciones_oeste_16_22['Equipo_pos'] + ' ' + df_posiciones_oeste_16_22['año'].astype(str) #añadimos el año a la columna equipo
df_posiciones_oeste_16_22 = df_posiciones_oeste_16_22.rename(columns={'Equipo_pos':'Team'})   #renombramos la columna equipo para poder juntarla posteriormente con la tabla general
del df_posiciones_oeste_16_22['año']   # borramos la columna año
df_posiciones_oeste_16_22
```

### 2.3. Unimos las dos tablas: ESTE y OESTE

```python
df_posiciones_16_22 = pd.merge(df_posiciones_este_16_22, df_posiciones_oeste_16_22,how='outer')
df_posiciones_16_22
```

### 2.4. Cargamos ahora los datos desde 2013 a 2015

```python
df_posiciones_13_15 = pd.read_csv('posiciones13-15.csv', sep=';')
df_posiciones_13_15
```

### 2.5. Unimos las tablas de 2013-2015 y 2016-2023

```python
# unimos las dos tablas
Tabla_total_posiciones = pd.merge(df_posiciones_13_15, df_posiciones_16_22 ,how='outer')
Tabla_total_posiciones
```

```python
# Eliminamos los datos de 2023 para unirlo mas tarde a la tabla general
Tabla_total_posiciones = Tabla_total_posiciones[~Tabla_total_posiciones['Team'].str.contains('2023')]
Tabla_total_posiciones
```

```python
Tabla_total_posiciones.to_csv("Tabla_total_posiciones.csv")
```

## 3. Unimos la tabla ofensiva con la de posicion y limpiamos

### 3.1. Unión

```python
df_NBA_posiciones = pd.merge(df_NBA, Tabla_total_posiciones, on = 'Team', how='left')
df_NBA_posiciones
```

```python
# vemos cuantos nulos hay y dónde estan 
df_NBA_posiciones[df_NBA_posiciones.isnull().any(axis=1)]
```

```python
#sustituimos los nulos por lo correspondiente
df_NBA_posiciones.loc[df_NBA_posiciones['Team'] == 'Oklahoma City Thunder 2014', 'posicion'] = 2
df_NBA_posiciones.loc[df_NBA_posiciones['Team'] == 'Oklahoma City Thunder 2013', 'posicion'] = 1
df_NBA_posiciones.loc[df_NBA_posiciones['Team'] == 'Memphis Grizzlies 2021', 'posicion'] = 8
df_NBA_posiciones.loc[df_NBA_posiciones['Team'] == 'New Orleans Pelicans 2022', 'posicion'] = 8
df_NBA_posiciones.loc[df_NBA_posiciones['Team'] == 'Atlanta Hawks 2022', 'posicion'] = 8
```

```python
# Volvemos a comprobar los nulos
df_NBA_posiciones[df_NBA_posiciones.isnull().any(axis=1)]
```

### 3.2. Limpieza

Quitamos el año de la columna Team

```python
for i in range(len(df_NBA_posiciones['Team'])):
    df_NBA_posiciones['Team'][i] = df_NBA_posiciones['Team'][i][:-4]
    
df_NBA_posiciones
```

Estudiamos la correlacion con la columna que nos interesa: CHAMP

```python
corr_matrix = df_NBA_posiciones.corr()
corr_matrix['CHAMP'].sort_values(ascending= False)
```

```python
from matplotlib import pyplot as plt
import seaborn as sn

df_NBA_posiciones_corta = df_NBA_posiciones[['PTS','AST','TOV','FGA','FG%','2P','3P','3P%']]
corr_matrix_corta = df_NBA_posiciones_corta.corr()
heatmap = sn.heatmap(corr_matrix_corta, annot=True)
heatmap.set(ylim=(0,len(corr_matrix_corta)))
heatmap.set(xlim=(0,len(corr_matrix_corta)))

plt.show()
```

Añadimos una nueva columna para saber la temporada, ya que lo hemos eliminado de la columna Team.
El año 2013 será la temporada 1

```python
df_NBA_posiciones['SEASON'] = (df_NBA_posiciones.index // 16) + 1
df_NBA_posiciones.tail(20)
```

Guardamos la tabla

```python
df_NBA_posiciones.to_csv("df_NBA_posiciones.csv")
```

## 4. Limpiamos la tabla defensiva

```python
df_NBA_defensive = pd.read_csv('df_NBA_defensive.csv', sep=',')
df_NBA_defensive.shape
```

```python
df_NBA_defensive = df_NBA_defensive.dropna()
df_NBA_defensive.shape
```

```python
df_NBA_defensive.tail(16)
```

```python
# Hacemos esto para eliminar las filas con Team = League Average
del df_NBA_defensive['Unnamed: 0']
filtro1 = df_NBA_defensive[df_NBA_defensive['Team'].str.contains('League Average')]
df_NBA_defensive = df_NBA_defensive.drop(filtro1.index)   # las borramos, ponemos index para indicar que es una fila
df_NBA_defensive.shape
```

```python
df_NBA_defensive
```

```python
df_NBA_defensive = df_NBA_defensive.reset_index(drop=True)
```

```python
df_NBA_defensive.tail()
```

```python
df_NBA_defensive.info()
```

```python
campeones = ['San Antonio Spurs 2014', 'Miami Heat 2013', 'Golden State Warriors 2015',
                    'Golden State Warriors 2017', 'Golden State Warriors 2018', 'Golden State Warriors 2022',
                    'Cleveland Cavaliers 2016', 'Toronto Raptors 2019', 'Los Angeles Lakers 2020',
                    'Milwaukee Bucks 2021']
df_NBA_defensive['CHAMP'] = 0
df_NBA_defensive.loc[df_NBA_defensive['Team'].isin(campeones), 'CHAMP'] = 1
```

```python
df_NBA_defensive[(df_NBA_defensive['CHAMP'] == 1 )]
```

```python
nueva_columnas = ['Opp ' + columna for columna in df_NBA_defensive.columns]
df_NBA_defensive = df_NBA_defensive.rename(columns=dict(zip(df_NBA_defensive.columns, nueva_columnas)))

df_NBA_defensive.columns
```

```python
df_NBA_defensive['SEASON'] = (df_NBA_defensive.index // 16) + 1
```

```python
df_NBA_defensive.columns
```

```python
#Renombramos las columnas Team, Champ, borramos las innecesarias y vemos los datos defensivos de 2023

df_NBA_defensive = df_NBA_defensive.rename(columns = {'Opp Team' : 'Team', 'Opp CHAMP':'CHAMP', 'Opp SEASON':'SEASON'})
df_NBA_defensive = df_NBA_defensive.drop(columns = 'Opp MP')
df_NBA_defensive = df_NBA_defensive.drop(columns = 'Opp G')

df_NBA_defensive.loc[df_NBA_defensive['Team'].str.contains('2023')]
```

```python
#borramos los datos de 2023 de la tabla defensiva.
filas_a_borrar = list(range(160,176))
df_NBA_defensive = df_NBA_defensive.drop(filas_a_borrar)
df_NBA_defensive.shape
```

```python
#estudiamos la correlacion de los datos defensivos
corr_matrix = df_NBA_defensive.corr()
corr_matrix['CHAMP'].sort_values(ascending= False)
```

```python
#Borramos el año de la columna Team 
for i in range(len(df_NBA_defensive['Team'])):
    df_NBA_defensive['Team'][i] = df_NBA_defensive['Team'][i][:-4]
        
df_NBA_defensive.head(5)
```

## 5. Unimos las tablas generales. Y limpiamos

### 5.1. Unimos

```python
Tabla_final_NBA = pd.merge(df_NBA_posiciones, df_NBA_defensive, on=['Team', 'SEASON'])
Tabla_final_NBA
```

## 5.2. Limpiamos

```python
del Tabla_final_NBA['CHAMP_y']
Tabla_final_NBA = Tabla_final_NBA.rename(columns = {'CHAMP_x':'CHAMP'})
Tabla_final_NBA.columns
```
